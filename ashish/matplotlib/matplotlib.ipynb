```python
import matplotlib.pyplot as plt

x = [0, 100, 200, 300, 400, 500, 600]
y = [0, 150, 150, 150, 150, 150, 0]

plt.plot(x, y)

plt.xlabel("Distance (km)")
plt.ylabel("Altitude (m)")
plt.title("Flight Trajectory from Hyderabad to Bangalore")

plt.show()
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
hours_slept = [7, 8, 6, 7, 9, 8, 9]

plt.bar(days, hours_slept)

plt.xlabel("Day of the Week")
plt.ylabel("Hours of Sleep")
plt.title("My Sleep Tracker")

plt.show()
```
