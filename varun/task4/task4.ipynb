```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg=sns.load_dataset("mpg")
mpg
```

scatterplot

```python
sns.scatterplot(x="mpg",y="displacement",data=mpg)
```

```python
sns.scatterplot(x="mpg",y="displacement",data=mpg,hue="weight")
```

joint plot


```python
sns.jointplot(x="mpg",y="displacement",data=mpg)
```

```python
sns.jointplot(x="mpg",y="displacement",data=mpg,kind="reg")
```

Box plot

```python
sns.boxplot(x="mpg",y="displacement",data=mpg)
```

pair plot

```python
sns.pairplot(mpg.select_dtypes(['number']),hue="acceleration")
```

correlation between the columns using heatmap

```python
sns.heatmap(mpg.corr(),annot = True,cmap="coolwarm")
```

```python
sns.heatmap(mpg.corr(),annot = True,cmap="icefire")
```
