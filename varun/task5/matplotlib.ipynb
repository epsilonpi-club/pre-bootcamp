```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sln
```

Hyetograph

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
precipitation = [0.5, 0.75, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5]
c=["#7B7385"]
 
plt.bar(months,precipitation,color=c,width=0.9)
plt.title("Hyetograph")
plt.xlabel("months")
plt.ylabel("precipitation")
```

Scatter plot



```python
longitude=np.random.rand(40)
latitude=np.random.rand(40)
plt.scatter(longitude,latitude,color="red",alpha=0.7)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.title("Scatter Plot")
```

```python
t = [0, 1, 2, 3, 4, 5] # time in seconds
x = [0, 10, 20, 30, 40, 50] # position in meters

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Motion of Car Moving in Straight Line')
plt.show()
```
