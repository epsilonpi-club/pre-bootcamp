```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall = [2, 4, 5, 8, 12, 6, 3, 10]
time_interval = [2, 4, 6, 8, 10, 12, 14, 16]
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(time_interval, rainfall, color='maroon')
ax.set_xlabel('Time (hours)')
ax.set_ylabel('Rainfall Intensity (mm/hour)')
ax.tick_params(axis='both', labelsize=10, colors='gray')
plt.title('Hyetograph showing rainfall intensity over time', fontsize=14, color='black')
plt.show()
```

```python
star_ra = [12.4436,	12.7958, 12.5194, 12.2272]
star_dec = [-60.3703, -59.6889, -57.1132,-58.7489]

fig, ax1 = plt.subplots()

ax1.scatter(star_ra, star_dec, s=100, c='gold', edgecolors='silver')

plt.title('The Four Brightest Stars in Crux', fontsize=16)
plt.xlabel('Right Ascension (hours)', fontsize=12)
plt.ylabel('Declination (degrees)', fontsize=12)
plt.show()
```

```python
def hill(x, max_h, peak, wid):
    return max_h * np.exp(-(x-peak)**2/(2*wid**2))

v = 1.0  
t = np.linspace(0, 10, 100)  
x = v * t  

max_h = 10.0  
peak = 0.0  
wid = 2.0  

y = hill(x, max_h, peak, wid)

fig, ax3 = plt.subplots()
ax3.plot(x, y, label='Hill', color='green')
ax3.plot(x, np.zeros_like(x), 'r--', label='Ground')
ax3.plot([0], [0], 'ko', label='Car')
ax3.legend()
ax3.set_xlabel('Position (m)')
ax3.set_ylabel('Height (m)')
ax3.set_title('Motion of a car crossing a hill')
plt.show()
```
