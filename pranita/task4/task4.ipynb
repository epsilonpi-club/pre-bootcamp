```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg=sns.load_dataset("mpg")
```

```python
mpg
```

```python
sns.scatterplot(x="mpg",y="cylinders",data=mpg,hue="weight")
```

```python
sns.pairplot(mpg.select_dtypes(['number']),hue="mpg")
```

```python
sns.jointplot(x="model_year",y="acceleration",data=mpg,kind="reg")
```

```python
sns.jointplot(x="model_year",y="acceleration",data=mpg,kind="hex")
```

```python
sns.boxplot(x="model_year",y="cylinders",data=mpg,hue="origin")
```

```python
sns.heatmap(mpg.corr(),annot = True,cmap="icefire")
```
