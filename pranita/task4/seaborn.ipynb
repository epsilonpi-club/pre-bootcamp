```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips)
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day")
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size")
```

```python
sns.histplot(tips['tip'],kde=True,bins=25)
```

```python
sns.distplot(tips['tip'],kde=True,bins=25)
```

```python
sns.displot(tips['tip'],kde=True,bins=25)
```

```python
sns.barplot(x="day",y="tip",data=tips,palette="YlGnBu")
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",dodge=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde",shade=True)
```
