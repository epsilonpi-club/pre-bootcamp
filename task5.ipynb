```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(60)

fig, ax4 = plt.subplots()
ax4.hist(data, bins=20, edgecolor='blue', color='black')

ax4.set_title('subset')
ax4.set_xlabel('Data Values')
ax4.set_ylabel('Finalset')

plt.show()
```

```python
star_c = [32.4436,	31.7958, 32.5194, 31.2272]
star_a = [-70.3703, -69.6889, -56.1132,-68.7489]

fig, ax1 = plt.subplots()

ax1.scatter(star_c, star_a, s=200, c='black', edgecolors='pink')

plt.title('the points', fontsize=14)
plt.xlabel('Right  ', fontsize=13)
plt.ylabel('Left ', fontsize=14)
plt.show()
```

```python
sunrise = [2, 4, 5, 8, 12, 6, 3, 10]
time_interval = [2, 4, 6, 8, 10, 12, 14, 16]
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(time_interval, sunrise, color='blue')
ax.set_xlabel('Time (hours)')
ax.set_ylabel('sun Intensity (mm/hour)')
ax.tick_params(axis='both', labelsize=10, colors='gray')
plt.title('Hyetograph showing sun intensity over time', fontsize=16, color='yellow')
plt.show()
```
